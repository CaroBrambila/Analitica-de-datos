```python
##definimos dos arreglos
import numpy as np
import matplotlib.pyplot as plt

ys = 200 + np.random.rand(100)
xs = np.arange(0, 100, 1)

#graficas los arreglos en segundo plano
plt.plot(xs, ys, '*')
plt.fill_between(xs, ys, 200, facecolor='red', alpha=0.6)

plt.title("Simple visualizaci√≥n")
plt.xlabel("x",size=16)
plt.ylabel("y",size=16)
plt.show()
```
